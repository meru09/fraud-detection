
Decision: The dataset contained 1,081 duplicate rows (1,062 legitimate and 19 fraudulent). We removed them to avoid over-representing identical observations during model training. This decision reduced the fraud class slightly (492 → 473), which is acceptable and documented.

### Class Distribution

After removing duplicate rows, the dataset contains:

- Legitimate transactions (Class = 0): 283,253 (99.834%)
- Fraudulent transactions (Class = 1): 473 (0.166%)

This confirms that the dataset remains extremely imbalanced. Therefore, accuracy alone will not be an appropriate evaluation metric. During model development, we will focus on metrics such as Precision, Recall, F1-score, and ROC-AUC.
